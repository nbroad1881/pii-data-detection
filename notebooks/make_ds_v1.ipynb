```python
from datasets import Dataset

# function to make random string of length n

def random_string(n):
    import string
    import random
    return ''.join(random.choice(string.ascii_lowercase) for _ in range(n))

ds = Dataset.from_parquet("/drive2/kaggle/pii-dd/data/ds-v2.pq")

ds = ds.rename_column("essay", "full_text")

ds = ds.add_column("document", [random_string(10) for x in range(len(ds))])
```

```python
import json

ds
```

```python
cols = ["document", "full_text", "tokens", "labels", "trailing_whitespace"]

data = []

for row in ds:
    data.append({col: row[col] for col in cols})
```

```python
data[0]
```

```python
with open("mixtral-8x7b-v1.json", "w") as fp:
    json.dump(data, fp)
```
