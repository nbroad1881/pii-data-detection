```python
import json
import os
from dotenv import load_dotenv
from pathlib import Path

dotenv_path = Path("../../.env")
if dotenv_path.exists():
    print("Loaded .env file!")
    load_dotenv(str(dotenv_path))


data = json.load(open(Path(os.environ["PROJECT_HOME_DIR"]) / "data/train.json"))
```

```python
doc = 12512

idx = -1
for i, d in enumerate(data):
    if d["document"] == doc:
        idx = i
        break

d = data[idx]
# swap "Jordi" with "SHIBPUR"
d["full_text"] = d["full_text"].replace("Jordi", "SHIBPUR")
idxs = [i for i, x in enumerate(d["tokens"]) if x == "Jordi"]

d["labels"] = [x if i not in idxs else "O" for i, x in enumerate(d["labels"])]
d["tokens"] = [x.replace("Jordi", "SHIBPUR") for x in d["tokens"]]

data[idx] = d
```

```python
import pandas as pd

tokens = []
labels = []
document = []
token_text = []

for d in data:
    tokens.extend(list(range(len(d["tokens"]))))
    token_text.extend(d["tokens"])
    labels.extend(d["labels"])
    document.extend([d["document"]] * len(d["tokens"]))

df = pd.DataFrame({"document": document, "token": tokens, "token_text": token_text, "label": labels})

df.to_parquet(Path(os.environ["PROJECT_HOME_DIR"]) / "data/corrected_train_v2.pq")
```
